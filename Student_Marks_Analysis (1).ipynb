Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

Create Data

```python
Students = np.array(["Tanisha", "Anushka", "Tanya", "Geetali", "Poonam"])
Marks = np.array([95, 88, 90, 72, 79])
```

Basic Analysis

```python
average = np.mean(Marks)
highest = np.max(Marks)
lowest = np.min(Marks)

print("Average Marks: ", average)
print("Highest Marks: ", highest)
print("Lowest Marks: ", lowest)
```

Find Topper

```python
topper_index = np.argmax(Marks)
topper_name = Students[topper_index]

print("Topper: ", topper_name)
```

Find students above average

```python
above_avg_students = Students[Marks > average]
print("Students above average: ", above_avg_students)
```

Assign grades

```python
grades = []
for m in Marks:
  if m >= 90:
    grades.append("A")
  elif m>= 75:
    grades.append("B")
  else:
    grades.append("C")
```

Display final report

```python
print("\n--- Student Report ---")
for i in range (len(Students)):
  print(Students[i], "|Marks:", Marks[i], "|Garde:", grades[i])
```

Bar graph

```python
plt.bar(Students, Marks)
plt.title("Student Marks Analysis")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

Line graph

```python
plt.plot(Students, Marks, marker='o')
plt.title("Marks Trend")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

Pie chart

```python
plt.pie(Marks, labels=Students, autopct='%1.1f%%')
plt.title("Marks Distribution")
plt.show()
```
